<h1>K Nearest Neighbor Algorithm applied to the IRIS-Dataset</h1>

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("Done")
```

<b>Spliting into Training and Testing Data 75 /25</b>

A typical first step in supervised learning is to split the available labeled data into training and test sets. Since the data was sorted by labels beforehand, it must first be shuffled to ensure that both sets have a representative distribution of the classes.

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split
X = iris_dataset['data']   
y = iris_dataset['target']  
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=3)
print('Data split')
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)

print(y_train)
```

```python
# TODO determine ratio test/train
```

<b>Create a model and train it</b>

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn
```

We created an instance of a learning algorithm which is stored in the variable knn. The orange box is the visual representation. 
The color orange indicates that is was not yet trained (it has not seen any data before). Lets look inside:

```python
knn.get_params()
# explanation of parameters at 
# https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
# e.g. metric:Minkowski+p:2 means euklidian distance is used
```

```python
# Lets train the algorithm now (Make it "fit" to the training data)
knn.fit(X_train, y_train)
```

The color changed to blue indicating that its now a trained algorithm, ready to make predictions with unlabeled examples.

<b>Making One Prediction manually</b>

Assume we found one iris in the wild and the leaf measurements taken are 5cm 2.9cm 1cm and 0.2cm. In the following we will first modify numpy arrays in a way to be able to illustrate how this new samples compares to existing samples.

```python
import numpy as np
X_predict = np.array([[5, 2.9, 1, 0.2]]) # one NEW example iris to predict
print("X_predict.shape:", X_predict.shape)
```

```python
# next we want to visualize the new point
# we are now concatenating the new iris to the training data
X_join = np.concatenate((X_train, X_predict), axis=0)
print(X_join[-10:]) # print the last 10
```

If you look at the last line of the previous cell, this is the new sample that we added!

```python
# creating a new label (3) for the new iris
y_join = np.concatenate((y_train, np.array([3])), axis=0)
print(y_join)
```

For the labels we now added the "3" as a new class besides the existing from 0-2

```python
#targetvalue 0 will be blue      -> setosa
#targetvalue 1 will be green     -> versicolor
#targetvalue 2 will be red       -> virginica
#targetvalue 3 -> Point to classify

import pandas as pd

import matplotlib as mpl

colmap = mpl.colors.ListedColormap(['b', 'g', 'r', 'c'])
join_dataframe = pd.DataFrame(X_join, columns=iris_dataset.feature_names)
allplots = pd.plotting.scatter_matrix(join_dataframe, c=y_join, 
                           figsize=(15, 15), diagonal='hist',
                           marker='o', hist_kwds={'bins': 20}, s=60,
                           alpha=0.8, cmap=colmap) #mglearn.cm3
```

```python
# Using the trained algorithm to make a prediction 
prediction = knn.predict(X_predict)
print("Prediction:", prediction)
print("Predicted target name:",
       iris_dataset['target_names'][prediction])
```

<b>Now running all test data</b>

The function *predict* operates on sets of samples, so we can make predictions for all Test-data with one single call. The results can then be compared to the ground-truth values.

```python
y_pred = knn.predict(X_test)
print("Test set predictions:\n", y_pred)
print("The original values were:\n", y_test)
```

❓Can you spot any difference between predicted values and groud truth labels ?



```python
# create a vector of comparison results
y_pred == y_test
```

```python
print(np.sum(y_pred == y_test))
```

```python
y_pred.size
```

```python
# how often was it right ? Lets coompute the percentage 
print("Test set score: {:.2f}".format(np.sum(y_pred == y_test)/y_pred.size))
```

```python
# the more general way is to call the score function of the algorithm
print("Test set score: {:.2f}".format(knn.score(X_test, y_test)))
```

For comparison we can also compute the training score, the quality of the predictions on the training data.

```python
print("Training set score: {:.2f}".format(knn.score(X_train, y_train)))
```
