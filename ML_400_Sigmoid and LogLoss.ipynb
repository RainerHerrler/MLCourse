# Sigmoid Experiments

**Plotting the pure sigmoid**

$h_\theta(x) \;=\; \frac{1}{1 + e^{-x}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-8,8,100)
y = 1/(1+math.e**-x)  

# Plot coordiante system
plt.plot(x,x*0,color='black')   # ground line
plt.plot(x,x*0+1,color='black', linestyle='dotted')
plt.plot(x,x*0+0.5,color='black', linestyle='dotted')
plt.plot((0,0),(0,1),color='black')    # vertical line

# plot the sigmoid
plt.plot(x,y)
plt.show()
```

**Plotting sigmoid varying $\theta_1$**

$\huge h_\theta(x)=\frac{1}{1+e^{-(\theta_0+\theta_1 x)}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Plot coordiante system
x = np.linspace(-8,16,100)
plt.plot(x,x*0,color='black')   # ground line
plt.plot(x,x*0+1,color='black', linestyle='dotted')
plt.plot((0,0),(0,1),color='black')    # vertical line


th0=0; # parameters to configure the sigmoid
th1=1;

for th in range(1,5):
    th1 = 1/th
    #th0 = th
    y = 1/(1+math.e**(-th0-(th1*x)))
    plt.plot(x,y,label=f'$\\theta_1={th1:.2f}$')

plt.legend(loc='lower right')
plt.show()    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Plot coordiante system
x = np.linspace(-8,16,100)
plt.plot(x,x*0,color='black')   # ground line
plt.plot(x,x*0+1,color='black', linestyle='dotted')
plt.plot((0,0),(0,1),color='black')    # vertical line


th0=0;  # parameters to configure the sigmoid
th1=1;

for th in range(0,5):
    #th1 = 1/th
    th0 = -th
    y = 1/(1+math.e**(-th0-(th1*x)))
    plt.plot(x,y,label=f'$\\theta_0={th0:.2f}$')

plt.legend(loc='lower right')
plt.show()    
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

x = np.linspace(-8,8,100)
y = 1/(1+math.e**-x)

plt.figure(figsize=(10, 6))
plt.plot(x,x*0,color='black')   # ground line
plt.plot(x,x*0+1,color='black', linestyle='dotted')
plt.plot((0,0),(0,1),color='black')    # vertical line

cost = -np.log(y)

plt.plot(x,y)
plt.plot(x,cost)
```
