**Creating meaningful vizualizations**

We learned pandas is the tool to show tabular data. Lets wrap the *ndarray* into the pandas-datastructure *DataFrame* and use it for displaying the numbers.

```python
import pandas as pd 
iris_dataframe = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)
display(iris_dataframe)
```

```python
print(iris_dataset['target'])
```

**Introducing the Scatter-Plot**

Four feature dimensions cannot be visualized in a two- or three-dimensional plot, but pandas provides a type of diagram that allows us to show pairwise relationships between all features. We will mix functionalities provided by matplotlib and pandas.

```python
#targetvalue 0 will be blue      -> setosa
#targetvalue 1 will be green     -> versicolor
#targetvalue 2 will be red       -> virginica
import matplotlib
cmap = matplotlib.colors.ListedColormap(['#0000aa', '#ff2020', '#50ff50']) #dirtyblue, brighterred, brightgreen
plots = pd.plotting.scatter_matrix(iris_dataframe, c=iris_dataset['target'], 
                           figsize=(10, 10), diagonal='hist',
                           marker='o', hist_kwds={'bins': 20}, s=40,
                           alpha=0.5, cmap=cmap)
```

We can see that when focusing on a pair of features, it is sometimes not possible to clearly distinguish between two classes. For example, if we only consider the sepal features (length and width), the green and red classes overlap in some regions, whereas the blue class is clearly separated. In contrast, when we focus on the lengths of both the sepal and petal, the data appears to be more easily separable.
