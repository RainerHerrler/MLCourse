# Lasso regression on the Extended Boston housing dataset

```python
# a lot of imports
%matplotlib inline
import sklearn
import mglearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
print("Done")
```

**Boston Extended**

<b>Lasso Regression - Introduction of L1-"Regularization"</b>

```python
from mglearn.datasets import load_extended_boston
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
X.shape
```

```python
# Lasso drives regularization to be coefficients to be exactly 0
from sklearn.linear_model import Lasso

lasso = Lasso().fit(X_train, y_train)  # no arguments means alpha = 1 
print("Training set score: {:.2f}".format(lasso.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso.coef_ != 0))
```

```python
lasso.alpha
```

```python
# we increase the default setting of "max_iter",
# otherwise the model would warn us that we should increase max_iter.
lasso001 = Lasso(alpha=0.01, max_iter=10000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso001.coef_ != 0))
```

```python
lasso00001 = Lasso(alpha=0.0001, max_iter=100000).fit(X_train, y_train)
print("Training set score: {:.2f}".format(lasso00001.score(X_train, y_train)))
print("Test set score: {:.2f}".format(lasso00001.score(X_test, y_test)))
print("Number of features used:", np.sum(lasso00001.coef_ != 0))
```

**Comparing Coefficients of Lasso with different $\alpha$**

```python
plt.figure(figsize=(12, 12))
plt.plot(lasso.coef_, 's', label="Lasso alpha=1 (0.21 Test Score)")
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01 (0.77 Test Score)")
plt.plot(lasso00001.coef_, 'v', label="Lasso alpha=0.0001 (0.64 Test Score)")


#plt.plot(ridge01.coef_, 'o', label="Best ridge")
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.legend()
plt.show()
```

**Lasso versus Ridge - Visualization of Coefficient Range**

```python
from sklearn.linear_model import Ridge
ridge01 = Ridge(alpha=0.1).fit(X_train, y_train)

plt.figure(figsize=(12, 12))
plt.plot(lasso001.coef_, '^', label="Lasso alpha=0.01 (0.77 Test Score)")
plt.plot(ridge01.coef_, 'o', label="Ridge alpha=0.1 (0.77 Test Score)")
plt.ylim(-25, 25)
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient magnitude")
plt.legend()
plt.show()
```
