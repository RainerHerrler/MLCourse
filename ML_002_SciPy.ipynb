```python
import numpy as np
from scipy import sparse
#create a 2D NumPy Eye Array 
eye = np.eye(4)
print(f"NumPy Array:\n{eye}")
```

<b>CSR Representation / Compressed sparse row<b>

```python
#convert the NumPy-Array into a SciPy sparse matrix in the CSR Format
#only the entries not equal to 0 are stored 
sparse_matrix = sparse.csr_matrix(eye)
print(f"SciPy sparse CSR matrix:\n{sparse_matrix}")
```

```python
sparse_matrix = sparse_matrix*5
sparse_matrix
```

```python
print(sparse_matrix)
```

<b>COO Representation - Coordinate Format<b>

```python
data = np.ones(4)
row_indices = np.arange(4)
col_indices = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_indices, col_indices)))
print("COO-Representation:\n{}".format(eye_coo))
```
