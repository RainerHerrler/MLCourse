# KNN for Regression with the California Housing Dataset

**Analysing the data**

```python
%matplotlib inline
import sklearn
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
print("done")
```

```python
#loading the dataset with skikit learn
housing = fetch_california_housing()
print("Data shape:", housing.data.shape)
print("Feature names:", housing.feature_names)
#print("Description:", housing.DESCR)

print("Labels", housing.target)
# do further analyis like inspect semantics of the columns etc
```

<b>Train and Evaluate a KNN</b><br>
Split the given housing data with ratio 75/25. Instantiate a KNeighborsRegressor with a neighborcount of 3. Fit the model with the training data.

```python
X_train, X_test, y_train, y_test = train_test_split(housing.data, housing.target, random_state=42)
print(X_train.shape)
print(X_test.shape)
```

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
```

Make the predictions for the test-set and calculate the error for all elements of the test data. Use the score-Funtion to calculate the R2-Score.

```python
y_predict = knn.predict(X_test);
print("Prediction  :", y_predict.round(decimals=2)[:20])
print("Ground Truth:", y_test.round(decimals=2)[:20])
print('-'*20)
print("Error       :", (y_test-y_predict).round(decimals=2)[:20])
print("")
```

```python
print("Training Score: ", knn.score(X_train, y_train))
print("Test Score: ", knn.score(X_test, y_test))
```

<b>Find the best n</b>    
Find the best n by checking multiple values and evaluating the R^2-Value. Is the result already good ? 

```python
for n_neighbors in [1, 3, 5, 9, 11, 13, 15]:
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    print("Neighbors\t", n_neighbors)
    print("Training Score\t", reg.score(X_train, y_train))
    print("Test Score\t",reg.score(X_test, y_test))
    print("-"*40)
```

<b>Improve Results by Rescaling/Normalizing the data</b> 
The KNN relies on the distance function (in 8 dimensional space in this case). The ranges and units are very different.
Next we try if we can improve results if we normalize the features. For this we can use the Class  StandardScaler. Below you find some example code how scaling can be used.

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train) # fits the Scaler with the training data
print("Skalierungsfaktoren ",scaler.scale_.shape) #contains the scaling factor for each feature
#print("Mittelwert aller Werte ",scaler.mean_.shape)
```

```python
X_train_scaled = scaler.transform(X_train) # applies the scaling to the given features
X_test_scaled = scaler.transform(X_test) # applies the scaling to the given features
```

```python
for n_neighbors in [1, 3, 5, 9, 11, 13, 15]:
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train_scaled, y_train)
    print("Neighbors\t", n_neighbors)
    print("Training Score\t", reg.score(X_train_scaled, y_train))
    print("Test Score\t",reg.score(X_test_scaled, y_test))
    print("-"*40)
```
