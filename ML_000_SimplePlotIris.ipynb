<b>With Pandas create a DataFrame to Plot it with a Scatter-Matrix</b>

```python
import pandas as pd 
iris_dataframe = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)
```

```python
from matplotlib import pyplot as plt
import numpy as np
my_function = lambda x: ['blue', 'green', 'red'][x]
my_vec_function = np.vectorize(my_function) 
colors = my_vec_function(iris_dataset['target'])
plt.rcParams["figure.figsize"] = (10,4)
myplot = plt.scatter(iris_dataset['data'][:,2],iris_dataset['target'],color=colors)


```

```python
plt.rcParams["figure.figsize"] = (10,1)
myplot = plt.scatter(iris_dataset['data'][:,2],iris_dataset['target']*0,color=colors)
```

```python
plt.rcParams["figure.figsize"] = (10,6)
myplot = plt.scatter(iris_dataset['data'][:,2],iris_dataset['data'][:,3],color=colors)
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
```

```python
plt.rcParams["figure.figsize"] = (10,6)
myplot = plt.scatter(iris_dataset['data'][:,2],iris_dataset['data'][:,3],color='blue')
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
```
