<b> Solving linear Equations </b><br>
$$ 4x + 3y + 2z = 25$$
$$-2x + 2y + 3z = -10$$
$$3x -5y + 2x = -4$$
<br>
$$
\begin{bmatrix}
  4 & 3 & 2\\
  -2 & 2 & 3\\
  3 & -5 & 2\\
\end{bmatrix} \cdot 
\begin{bmatrix}
  x \\
  y \\
  z \\
\end{bmatrix}
=
\begin{bmatrix}
  25 \\
  -10 \\
  -4 \\
\end{bmatrix}
$$

```python
A = np.array([[4, 3, 2], [-2, 2, 3], [3, -5, 2]])
b = np.array([25, -10, -4])
Ainv = np.linalg.inv(A)
xvec = np.linalg.inv(A).dot(b)
print("Solution (x,y,z)=", xvec)
#Check
print("Check:", A.dot(xvec))

#Often the better Solution
print("Solution (x,y,z) ohne inverse=", np.linalg.solve(A,b))
```
