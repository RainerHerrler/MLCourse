<b>Required Imports</b>

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.neighbors import KNeighborsClassifier
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics.pairwise import euclidean_distances
from sklearn.utils import check_random_state, shuffle


print('sklearn:', sklearn.__version__)
```

**Creating the datsset**

```python
# data samples defined  as in https://github.com/amueller/introduction_to_ml_with_python/blob/main/mglearn/plot_knn_classification.py

g = check_random_state(4)

# fixed setup (2 centers, 2 features, std=1)
centers = g.uniform(-10, 10, size=(2, 2))
n = [15, 15]  # 30 samples split evenly

X = np.vstack([centers[i] + g.normal(scale=1.0, size=(n[i], 2)) for i in range(2)])
y = np.array([i for i in range(2) for _ in range(n[i])])

X, y = shuffle(X, y, random_state=g)

# apply your modifications
y[[7, 27]] = 0
mask = np.ones(len(X), bool)
mask[[0, 1, 5, 26]] = 0
X, y = X[mask], y[mask]


# Creating three test samples
X_test = np.array([[8.2, 3.66214339], [9.9, 3.2], [11.2, .5]])
```

<b>Visualizing the data</b>

```python
markers = ['o', '^']
for i, class_value in enumerate(np.unique(y)):
    plt.scatter(
        X[y == class_value, 0],
        X[y == class_value, 1],
        marker=markers[i % len(markers)],
        s=100,                    # 👈 größere Punkte
        edgecolor='black',        # 👈 schwarzer Rand
        label=f"Class {class_value}"
    )

#uncomment to see test data
#plt.scatter(X_test[:,0],X_test[:,1], marker="*",s=100,edgecolor="black",label="new",c="lightgreen")

plt.legend(loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.show()

print("X.shape:", X.shape)
```

<b> Vizualize effects of different parameters n = 1, 3, 5</b>

```python
from sklearn.neighbors import KNeighborsClassifier

# example value - set this as needed
n_neighbors = 1

# draw arrows from each test point to its k nearest neighbors
dist = euclidean_distances(X, X_test)
closest = np.argsort(dist, axis=0)
for x, neighbors in zip(X_test, closest.T):
    for neighbor in neighbors[:n_neighbors]:
        plt.arrow(x[0], x[1],X[neighbor, 0] - x[0],X[neighbor, 1] - x[1],
            head_width=0,fc='k',ec='k',length_includes_head=True
        )

# fiting the algorithm
clf = KNeighborsClassifier(n_neighbors=n_neighbors)
clf.fit(X, y)

# create predictions for test points
y_test_pred = clf.predict(X_test)

# plot training data
plt.scatter( X[y == 0, 0], X[y == 0, 1],marker='o', s=100, edgecolor='black',label='training 0')
plt.scatter(X[y == 1, 0], X[y == 1, 1],marker='^', s=100, edgecolor='black',label='training 1')

# plot test points with predicted class
plt.scatter(X_test[y_test_pred == 0, 0], X_test[y_test_pred == 0, 1],marker='*', s=200, edgecolor='black',label='test 0')
plt.scatter(X_test[y_test_pred == 1, 0], X_test[y_test_pred == 1, 1],marker='*', s=200, edgecolor='black',label='test 1')

plt.xlabel("feature 0")
plt.ylabel("feature 1")
plt.legend(loc="best")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for n_neighbors, ax in zip([1, 3, 9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X, y)
    plot_decision_regions(X, y,clf=clf,ax=ax, scatter_kwargs={'s': 100});
    ax.set_title(f"{n_neighbors} neighbor(s)")
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

plt.show()
```
