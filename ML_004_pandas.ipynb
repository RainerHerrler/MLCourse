<b>Also good for showing numpy Arrays</b>

```python
# creating a numpy array with 6*4
np.random.seed(seed=2022)
X = np.random.randint(100, size=(6,4))+1 # Features 
y = np.random.randint(2, size=6) #labels
print('Labels', y, sep='\n')
print('Features',X, sep='\n')
Xy = np.c_[X,y] # concatenates a column at the end of a matrix
print('Features&Lables',Xy,sep='\n')
```

```python
# Creating Line and Column Labels
myindex = ["Example {}".format(x) for x in list("123456")]
featureNames = ["Feature {}".format(x) for x in list("ABCD")]
featureNames.append('Label')
df = pd.DataFrame(Xy, index=myindex, columns=featureNames)
display(df)
```
