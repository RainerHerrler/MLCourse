# Linear Regression & Cost function

**Analysing the data** (Artificial Wave Dataset)

```python
%matplotlib inline
import sklearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
print(sklearn.__version__)
```

```python
# We hhad used this before already
# this is the funnction for data creation / function from https://github.com/amueller/mglearn/blob/main/mglearn/datasets.py
def make_wave(n_samples=100):
    rnd = np.random.RandomState(42)
    x = rnd.uniform(-3, 3, size=n_samples)
    y_no_noise = (np.sin(4 * x) + x)
    y = (y_no_noise + rnd.normal(size=len(x))) / 2
    return x.reshape(-1, 1), y
print("Function make_wave defined!")
```

```python
x,y = make_wave(n_samples=40)
plt.plot(x,y,'o')           # plotting original points (in blue)
plt.plot(x, (x*0.48)-0.06)  # plottinhg our model/hypothesis
plt.show()
```

```python
# Plotting two hypothesis functions
#def f1(x):
#    return (x*0.48)-0.06

x,y = mglearn.datasets.make_wave(n_samples=40)
h1 = lambda x : (x*0.48)-0.06
h2 = lambda x : (x*0.18)-1.56

plt.plot(x,y,'o')           # plotting original points

plt.plot(x, h1(x), label='h1')  # plottinhg our model/hypothesis
plt.plot(x, h2(x), label='h2')  # plottinhg our model/hypothesis
plt.legend()
plt.show()
```

### Plot Cost Function in 3D
now plot the cost function in 3D for the given 40 points based on the formula

$J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} \left( h_\theta(x[i]) - y[i] \right)^2$


```python
# defining the cost function J
def J(th0, th1,x,y):
  return np.sum((y-((x*th1)+th0))**2)/(2*len(y))

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(np.arange(-2, 2, 0.1), np.arange(-2, 2, 0.1))
zs = np.array([J(th0,th1,x,y) for th0,th1 in zip(X.flatten(), Y.flatten())])
Z = zs.reshape(X.shape)
ax.plot_surface(X, Y, Z)

ax.set_xlabel("$\\theta_0$")
ax.set_ylabel('$\\theta_1$')
ax.set_zlabel('$J(\\theta_0,\\theta_1)$')
plt.show()
```

```python
# Another way to show this is a contour plot that draws
# level curves of the cost function J(θ0, θ1), where each line represents points with equal error
plt.contour(X, Y, Z,50)
plt.xlabel("$\\theta_0$")
plt.ylabel('$\\theta_1$')
plt.scatter(0.06, 0, color='orange', s=250)
plt.show()
```

```python
# To illustrate how close the hypothsis functions are to the optimum
plt.contour(X, Y, Z,30)
plt.xlabel("$\\theta_0$")
plt.ylabel('$\\theta_1$')

# now add the two results for h1 and h2
# h1 = lambda x : (x*0.48)-0.06  # orange 
# h2 = lambda x : (x*0.18)-1.56  # green

plt.scatter(0.06, 0.48, color='orange', s=20)
plt.scatter(1.56, 0.18, color='green', s=20)
plt.show()
```
