<b>Matplotlib with builtin python lists</b><br><br>
Suppose we want to plot $sin(x)$ and $cos(x)$ between $[-10,10]$ with a resolution 0.2 (stepsize) 

```python
from math import sin, cos
r=range(-100,101,2) #range does just work with integer

# TODO
# output len and contents
```

```python
# here you see something new.. this is called "list comprehension"
x = [x/10 for x in r]
y = [sin(y) for y in x]
z = [cos(z) for z in x]
len(x)
```

```python
#Alternative/Equivalent to the list comprehension syntax:
list(map(lambda x: x/10, r))
```

```python
#Plotting the data (2 curves in one graph)
import matplotlib.pyplot as plt
plt.plot(x,y) # blue
plt.plot(x,z) # orange
plt.show()
```

<b>Matplotlib with numpy</b>

```python
x= np.linspace(-10,10,101)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,marker="o", c='red')
plt.plot(x,z,marker="x")
plt.show()
```

<b>Area Scatter Plot</b>

API description of the Scatter Plot
https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html

```python
import math
np.random.seed(19680801)

N = 10
x = np.random.rand(N)*100
y = np.random.rand(N)*100
colors = np.random.rand(N)
#colors
area = math.pi*(10 * np.random.rand(N))**2  # 0 to 10 point radii
plt.scatter(x, y, s=area, c=colors, alpha=0.5) 
plt.show()
```

<b>Drawing Multiple Plots</b>

```python
plt.scatter(x, y, c='blue', marker='x')
plt.scatter(x, y, s=area, c=colors, alpha=0.5) 
plt.show()
```

<b>Arranging multiple Plots horizontally</b>

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x, y)
ax2.plot(x, -y)
plt.show()
```

<b>3-Dimensional plots</b>

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111, projection='3d')

# Grab some test data.
X, Y, Z = axes3d.get_test_data(0.04)
ax.plot_surface(X, Y, Z, cmap='viridis' , antialiased=True)
plt.show()
Z.shape
```
