# KNN for Regression

**Analysing the data** (Artificial Wave Dataset)

```python
%matplotlib inline
import sklearn
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
print("Sklearn version", sklearn.__version__)
```

```python
# this is our new funnction for data creation / function from https://github.com/amueller/mglearn/blob/main/mglearn/datasets.py
def make_wave(n_samples=100):
    rnd = np.random.RandomState(42)
    x = rnd.uniform(-3, 3, size=n_samples)
    y_no_noise = (np.sin(4 * x) + x)
    y = (y_no_noise + rnd.normal(size=len(x))) / 2
    return x.reshape(-1, 1), y
```

```python
X,y = make_wave(n_samples=40)
plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel("Feature")
plt.ylabel("Target Value")
plt.show()
print('Done')
```

```python
## Possible model 1
plt.plot(X,y,'o')           # plotting original points

plt.plot(X, (X*0.48)-0.06)  # plotting our model/hypothesis
plt.show()
```

```python
#Possible model 2
plt.plot(X,y,'o') 
Xsorted, Ysorted = (list(t) for t in zip(*sorted(zip(X, y)))) # using argument unpacking
plt.plot(Xsorted, Ysorted)
plt.show()

```

**Understand the underlying structure of the data**

```python
# the real underlying structure becomes visible if we increase the sample size  
X,y = make_wave(n_samples=4040)
plt.plot(X,y,'o', alpha=0.2)
plt.ylim(-3,3)
plt.xlabel("Feature")
plt.ylabel("Target Value")
plt.show()
print('Done')
```

<b>How to do KNN Regression in SKLearn</b>

```python
from sklearn.neighbors import KNeighborsRegressor
X, y = make_wave(n_samples=40)

# split the wave dataset into a training and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# instantiate the model and set the number of neighbors to consider to 3
reg = KNeighborsRegressor(n_neighbors=3)
# fit the model using the training data and training targets
reg.fit(X_train, y_train)
```

```python
#dir(reg)
#print(reg._fit_X, reg.y)
```

```python
# Make predictions and show
y_predict = reg.predict(X_test);
print("Test set predictions:", y_predict.round(decimals=2))
print("Ground Truth        :", y_test.round(decimals=2))
```

<b> How good was that ? We need to calculate a score </b>

```python
print("Error               :", np.abs((y_predict-y_test)).round(decimals=2))
```

```python
print("Mean Absolute Error :", np.sum(np.abs((y_predict-y_test))).round(decimals=2))
```

```python
print("Test set R2-Score   : {:.2f}".format(reg.score(X_test, y_test)))
```

<b> Again test R^2 (Quality) for different n </b>

```python
for n_neighbors in [1, 3, 9]:
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    print("Neighbors\t", n_neighbors)
    print("Training Score\t", reg.score(X_train, y_train))
    print("Test Score\t",reg.score(X_test, y_test))
    print("-"*40)
    
```

<b> Plot it for a better understanding </b>

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
# create 1,000 data points, evenly spaced between -3 and 3
line = np.linspace(-3, 3, 1000).reshape(-1, 1) ## -1 => 1000 
for n_neighbors, ax in zip([1, 3, 9], axes):
    # make predictions using 1, 3, or 9 neighbors
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train, y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', color='#0000aa', markersize=8)
    ax.plot(X_test, y_test, 'v', color='#ff2020', markersize=8)
    ax.set_title(
        "{} neighbor(s)\n train score: {:.2f} test score: {:.2f}".format(
            n_neighbors, reg.score(X_train, y_train),
            reg.score(X_test, y_test)))
    ax.set_xlabel("Feature")
    ax.set_ylabel("Target")
axes[0].legend(["Model predictions", "Training data/target","Test data/target"
               ], loc="best")
plt.show()
```
