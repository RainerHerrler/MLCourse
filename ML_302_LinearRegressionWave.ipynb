# Linear regression on Wave</b>

```python
# a lot of imports
%matplotlib inline
import sklearn
import numpy as np
import matplotlib.pyplot as plt
#from sklearn import datasets
#from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from mglearn.datasets import make_wave
print(sklearn.__version__)
```

```python
# We hhad used this before already
# this is the funnction for data creation / function from https://github.com/amueller/mglearn/blob/main/mglearn/datasets.py
def make_wave(n_samples=100):
    rnd = np.random.RandomState(42)
    x = rnd.uniform(-3, 3, size=n_samples)
    y_no_noise = (np.sin(4 * x) + x)
    y = (y_no_noise + rnd.normal(size=len(x))) / 2
    return x.reshape(-1, 1), y
print("Function make_wave defined!")
```

**Creating and visualizing the data**

```python
from sklearn.linear_model import LinearRegression

X, y = make_wave(n_samples=60)

plt.figure(figsize=(6, 4))
plt.plot(X, y, 'o')
plt.grid()
plt.show()
```

**Splitting and Fitting**

```python
from sklearn.linear_model import LinearRegression

X, y = make_wave(n_samples=60)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

lr = LinearRegression().fit(X_train, y_train)
lr
```

```python
# what was learned by the model ? 
print(lr.coef_[0], lr.intercept_)
```

**Joint Visualisation**

```python
lineX = np.linspace(-3, 3, 100).reshape(-1, 1)
liney = lineX*lr.coef_[0]+lr.intercept_ 
#lineY = lr.predict(line) # that would be the same

plt.figure(figsize=(6, 4))
plt.plot(X, y, 'o')
plt.plot(lineX, liney)
plt.ylim(-3, 3)
plt.legend(["training data", "model"], loc="best")
plt.grid(True)
plt.show()
```

### Evaluate the result

```python
training_score = lr.score(X_train, y_train)
test_score     = lr.score(X_test, y_test)
print("Training set score: {:.2f}".format(training_score))
print("Test set score: {:.2f}".format(test_score))
```

Very simple model. Training and test score close together. Underfitting!
